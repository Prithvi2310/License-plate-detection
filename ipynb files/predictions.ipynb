```python
import numpy as np
import tensorflow as tf
import plotly.express as px
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
model = tf.keras.models.load_model('../models/object-detection.h5')
```

Testing the model for 1 image

```python
path = '../Images/images/Cars0.png'
```

```python
image = load_img(path)
```

```python
image = np.array(image, dtype=np.uint8) # 8 bit array (0,255)
```

```python
image1 = load_img(path, target_size=(224,224))
```

```python
image_arr_224 = img_to_array(image1)/255
```

```python
h, w, d = image.shape
```

```python
print('Height of the image = ',h)
print('Width of the image = ', w)
```

```python
fig = px.imshow(image)
fig.update_layout(width=500, height=300, margin=dict(l=10, r=10, b=10, t=10))
```

```python
#shape
image_arr_224.shape
```

In order to pass this image to the model, we need to provide the data in a dynamic fourth dimension. The extra dimension indicated the number of images being provided which is 1 in this case.

```python
test_arr = image_arr_224.reshape(1,224,224,3)
test_arr.shape
```

#### Test Prediction 

```python
coords = model.predict(test_arr)
```

```python
coords
```

#### Denormalizing the output
The output we got is normalized. To visualize the output we have to denormalize it. So what we need to do is convert them back to coordinates using the original image dimensions.

```python
denorm = np.array([w,w,h,h])
coords = coords * denorm
coords
```
